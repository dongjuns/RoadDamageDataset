```python
import os
import xml.etree.ElementTree as ET
from shutil import copyfile
```

```python
labels = ["D00", "D10", "D20", "D40"]

annoFiles = os.listdir(os.path.join(os.getcwd(), "Annotations/"))
jpgFiles = os.listdir(os.path.join(os.getcwd(), "JPEGImages/"))

print(len(annoFiles), len(jpgFiles))
```

```python
new_train_annoFiles = os.listdir(os.path.join(os.getcwd(), "new_train/Annotations/"))
new_val_annoFiles = os.listdir(os.path.join(os.getcwd(), "new_val/Annotations/"))

print(len(new_train_annoFiles), len(new_val_annoFiles))
```

```python
# making training dataset

#for annoFile in annoFiles[:n]:
for annoFile in annoFiles:
    print(annoFile)
    print(annoFile.split(".")[0])
    tree = ET.parse(os.path.join(os.getcwd()+"/Annotations/", annoFile))
    root = tree.getroot()
    for obj in root.findall("object"):
        a = obj.find("name").text
        print("here is ", a)
        if a not in labels:
            print(a)
            root.remove(obj)
    
    if len(root.findall("object")) > 0:
        tree.write("new_train/Annotations/" + annoFile)
        copyfile(os.path.join("JPEGImages/", annoFile.split(".")[0])+".jpg", "new_train/JPEGImages/"+annoFile.split(".")[0]+".jpg")
    else:
        print("oops")
```

```python
# making validation dataset

#for annoFile in annoFiles[n:]:
for annoFile in annoFiles[7000:10000]:
    print(annoFile)
    print(annoFile.split(".")[0])
    tree = ET.parse(os.path.join(os.getcwd()+"/Annotations/", annoFile))
    root = tree.getroot()
    for obj in root.findall("object"):
        a = obj.find("name").text
        print("here is ", a)
        if a not in labels:
            print(a)
            root.remove(obj)
    
    if len(root.findall("object")) > 0:
        tree.write("new_val/Annotations/" + annoFile)
        copyfile(os.path.join("JPEGImages/", annoFile.split(".")[0])+".jpg", "new_val/JPEGImages/"+annoFile.split(".")[0]+".jpg")
    else:
        print("oops")
```

```python
new_train_annoFiles = os.listdir(os.path.join(os.getcwd(), "new_train/Annotations/"))
new_val_annoFiles = os.listdir(os.path.join(os.getcwd(), "new_val/Annotations/"))

print(len(new_train_annoFiles), len(new_val_annoFiles))
```
